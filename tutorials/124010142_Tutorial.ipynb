```python
'''
Author: Vardhini K
Date: 25 Aug 2023
Time: 11:30 AM
Create a dictonary containing numbers as keys and squares as values upto user input number
'''
#Input n
n = int(input('Enter the value of n : '))

#Create Dictonary
square_dict = {}

#Loop for Dictonary
for i in range(1, n+1):
  square_dict[i] = i*i
  i+=1

print('The Output Dictonary is : ', square_dict)
```
