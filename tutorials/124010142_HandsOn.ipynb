```python
#Import Regex Library
import re

#Print heading
print("Gene\t\t\t\tMotif Count\t\tMotif")
print('----------------------------------------------------------------')

#Define function to find Motifs
def find_pattern(pattern, sequence):
  return re.findall(pattern, sequence)

#Open the Genome Sequence
with open('grape_promoters.txt', 'r') as fh:
  #Read entire file
  whole_sequence = fh.readlines()
#Spilt the file to lines
for i in range(0, len(whole_sequence)):
  gene_list = whole_sequence[i].split(' ')
  #Declare gene and sequence
  gene = gene_list[0]
  sequence = gene_list[-1]
  #Declare the required motifs in Regex
  pattern_list = ['[AT]GATA[GA]', '[GCT]ACGTG[GT][AC]', 'TTGAC']
  #Call the function
  for pattern in pattern_list:
    motif = find_pattern(pattern, sequence)
    seen = set()
    #Count no. of occurances
    for m in motif:
      if m not in seen:
        count = motif.count(m)
        #Print the motif along with count
        print(f'{gene}\t\t{count}\t\t\t{m}')
        seen.add(m)
```
