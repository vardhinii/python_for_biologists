##***Lab 4 - Hydropathy plot of a protein sequence, and uses in identifying the transmembrane segment through visualization in matplotlib***

*Aim: In this exercise, you will learn how to develop a Python code for a hydropathy plot of a protein sequence*

```python
'''
Author: Vardhini K
Date: 23 Aug 2023
Time: 9:30 AM
This code takes a hydropathy plot for a given protein sequence
'''
# Import matplotlib class pyplot as plt (alias)
import matplotlib.pyplot as plt

# Input of FASTA file
fasta_file = input('Enter your fasta file: ')

fasta_header = ""
protein_sequence = ""

# Open the fasta file and seperate the sequence
with open(fasta_file, "r") as fh:
    for line in fh:
        line = line.strip()
        if line.startswith('>'):
            fasta_header = line[1:]
        else:
            protein_sequence += line

# Define the hydropathy values
hydropathy = { 'I':4.5, 'V':4.2, 'L':3.8, 'F':2.8, 'C':2.5, 'M':1.9, 'A':1.8, 'G':-0.4, 'T':-0.7, 'S':-0.8, 'W':-0.9, 'Y':-1.3, 'P':-1.6, 'H':-3.2, 'E':-3.5, 'Q': -3.5, 'D':-3.5, 'N':-3.5, 'K':-3.9, 'R':-4.5 }

# Compute hydropathy scores for each residue in the sequence
hydropathy_scores = [hydropathy[residue] for residue in protein_sequence]

# Calculate the moving average over a window of size w
w = int(input('Enter your moving window: '))

# Calculation of moving average
conv_scores = [sum(hydropathy_scores[i:i+w]) / w for i in range(len(hydropathy_scores) - w + 1)]

# Create the hydropathy plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(conv_scores)+1), conv_scores, 'm-o')
plt.xlabel('Residue Number')
plt.ylabel('Hydropathy Score')
plt.title('Hydropathy Plot of Protein Sequence')
plt.grid(True)
plt.show()
```
